## **Phase 2:**

#### **Data Mining**
For this phase, we will uncover any hidden patterns, trends, and insight in the **Eye Cancer** dataset. With this, we will consider using clustering as a primary technique to identify any distinct groups of patients that has similar characteristics. Since data preprocessing has already been made, we can proceed with the desired data mining techniques.

Let's observe and plot all patients based on `Age` and `Survival_Time_Months` into 4 groups. We will use the `kmeans.py` to obtain the plot.

```python
from kmeans import KMeans

# Select numeric features for clustering (e.g., Age and Survival_Time_Months)
X = eye_cancer_df[['Age', 'Survival_Time_Months']].dropna().reset_index(drop=True)
kmeans = KMeans(4, 0, 2, len(X), X)

# label clusters to the DataFrame
centroids = kmeans.initialize_centroids(X)
groups = kmeans.train(X, 100)

plt.figure(figsize=(10,6))
colors = ['r', 'g', 'b', 'purple']

for i, color in enumerate(colors):
    mask = (groups == i)
    subset = X[mask]
    plt.scatter(subset['Age'], subset['Survival_Time_Months'], c=color, label=f'Cluster {i}', alpha=0.6)

plt.scatter(kmeans.centroids['Age'], kmeans.centroids['Survival_Time_Months'], c='k', marker='*', s=200, label='Centroids')

plt.xlabel('Age')
plt.ylabel('Survival Time (Months)')
plt.legend()
plt.grid()
plt.title('Eye Cancer Patients Clustering')
plt.show()

print("Centroids:\n", kmeans.centroids.round(2))
```

Based on the data above, we observed four distinct clusters:
- Cluster 0: Older patients (approx. 67 years of age) are grouped with longer survival times (92 months)
- Cluster 1: Younger patients (approx. 23 years of age) are grouped with short survival times (approx. 31 months)
- Cluster 2: Older patients (approx. 69 years of age) are grouped with short survival times (approx. 32 months)
- Cluster 3: Younger patients (approx. 23 years of age) are grouped with long survival times (approx. 90 months)

Let's observe the number of patients for each clusters

```python
#get the number of patients in each cluster
cluster_counts = pd.Series(groups).value_counts().sort_index()
print("Number of patients in each cluster:")
for cluster, count in cluster_counts.items():
    print(f"Cluster {cluster}: {count} patients")
```

> The clustering reveals that **Cluster 1** (Young with short survival) tends to have more patients, while the rest of the clusters almost have similar counts. However, we can observe more besides the number of patients based on survival.

**Let's encode categorical variables and apply `kmeans.py` to group patients based on `Age`, `Stage_at_Diagnosis`, and `Treatment_type`.**

For this line, we will import `sklearn` to use `LabelEncoder`. This is usable for categorical text data to convert into numeric labels.

```python
from sklearn.preprocessing import LabelEncoder

# Encoding the categorical values
eye_cancer_df['StageDiagnosis'] = LabelEncoder().fit_transform(eye_cancer_df['Stage_at_Diagnosis'])
eye_cancer_df['TreatmentType'] = LabelEncoder().fit_transform(eye_cancer_df['Treatment_Type'])

# Select features for clustering
X_cluster = eye_cancer_df[['Age', 'StageDiagnosis', 'TreatmentType']].dropna().reset_index(drop=True)

kmeans = KMeans(4, 0, X_cluster.shape[1], len(X_cluster), X_cluster)
centroids = kmeans.initialize_centroids(X_cluster)
groups = kmeans.train(X_cluster, 100)

# Add cluster labels to DataFrame
eye_cancer_df['Patient_Cluster'] = -1
eye_cancer_df.iloc[:len(groups), eye_cancer_df.columns.get_loc('Patient_Cluster')] = groups
```

Now, we can check the correlation with the `Outcome_Status` to see if these clusters actually correspond to meaningful differences in patient outcomes.

```python
# Check correlation with Outcome_Status
cluster_outcome = eye_cancer_df.groupby('Patient_Cluster')['Outcome_Status'].value_counts(normalize=True).unstack().fillna(0)
print(cluster_outcome)

plt.figure(figsize=(8,6))
sns.countplot(data=eye_cancer_df.iloc[:len(groups)], x='Patient_Cluster', hue='Outcome_Status')
plt.title('Cluster vs Outcome Status')
plt.xlabel('Cluster')
plt.ylabel('Patient Count')
plt.legend(title='Outcome Status')
plt.tight_layout()
plt.show()
```

Let's observe these results from here with a table:
| Cluster | Age Group | Survival Time | Active % | Deceased % | In Remission % |
|---|---|---|---|---|---|
| 0 | Older (~67) | Long (92mo) | 31.0% | 34.4% | 34.5% |
| 1 | Younger (~23) | Short (31mo) | 35.2% | 32.4% | 32.3% |
| 2 | Older (~69) | Short (32mo)	| 30.6% | 35.8% | 33.5% |
| 3 | Younger (~23) | Long (90mo) | 32.2% | 34.1% | 33.7% |

> These insights have little meaningful outcomes due to all clusters show nearly identical distributions (~31-35% per outcome). With this, the features of `Age` and `Survival_Time_Months` might not correlate strongly with `Outcome_Status`.

**What if we consider features with stronger clinical relevance?**

**Association Rule Mining**

Let's discover patterns on the dataset with the use of Association Rule Mining. With this, let's utilize the Apriori algorithm to uncover frequent patterns and associations between patient attributes and survival outcomes.

**What is Apriori algorithm?**
This is one of the prominent data mining technique where it is used to uncover relationships and dependencies between items in a large dataset. 
With the given technique, we will plan a goal of an "if-then" rules that may reveal patterns in the eye cancer dataset.

In this case, let's understand how combination of features such as `Age` and `Stage_at_Diagnosis` can relate to survival outcomes.

For this, we will need a library called **mlxtend**, it is a python library to utilize `apriori` and `association_rules` to get the frequent itemsets in the dataset and to generate an if-then rules from the itemsets.

```python
from mlxtend.frequent_patterns import apriori, association_rules

eye_cancer_df['Age_Group'] = pd.cut(eye_cancer_df['Age'], bins=[0, 40, 60, 120], labels=['<40', '40-60', '>60'])
eye_cancer_df['Survival_Group'] = pd.cut(eye_cancer_df['Survival_Time_Months'], bins=[0, 12, 24, eye_cancer_df['Survival_Time_Months'].max()], labels=['<12', '12-24', '>24'])

#convert categorical variables to one-hot encoding
apriori_df = pd.get_dummies(eye_cancer_df[['Age_Group', 'Stage_at_Diagnosis', 'Survival_Group']])

#Find the frequent itemsets and association rules
frequent_itemsets = apriori(apriori_df, min_support=0.05, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)

survival_rules = rules[rules['consequents'].apply(lambda x: any('Survival_Group' in str(i) for i in x))]
print(survival_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

Let's provide any notable trends and rules:
1. Most rules predict longer survival times (>24 months) as the consequent, indicating that the dataset may skew toward better-outcome patients.
2. Rules have high confidence (80-83%), meaning when the antecedent occurs, the consequent will like happen.
3. Lift values have a higher result, suggesting that these relationships are only slightly stronger than random chance, which indicates weak associations.

**Insights:**
1. All age groups have similar confidence (~80-81%) for predicting `24 months` survival, but this might contradicts for older age since they typically have a risk factor. The rules show no negative correlation.
2. Stage III has the highest confidence (81.5%) for long survival, but even slightly better than Stage I (80.6%). Although, clinical results might contradict typical oncology patterns like earlier stage mostly have better prognosis.

#### **Statistical Inference 1:** *Inference for Multiple Means of Survival Time by Cancer Type using ANOVA*

Recall from our EDA 1, that the variables between the cancer types only varies slightly.

Let's dig further and use ANOVA to compare the means between the 3 cancer types.

```python
#Prepare data for multiple means (mm)
lymphoma_survival_mm = lymphoma_survival_df.agg(['mean', 'std','count'])
melanoma_survival_mm = melanoma_survival_df.agg(['mean', 'std','count'])
retinoblastoma_survival_mm = retinoblastoma_survival_df.agg(['mean', 'std','count'])

# Combine
type_survival_mm = pd.DataFrame({
    'Lymphoma': lymphoma_survival_mm,
    'Melanoma': melanoma_survival_mm,
    'Retinoblastoma': retinoblastoma_survival_mm
})

type_survival_mm
```

The Core Concept for ANOVA states that if the means are the same, then their variability should be quite low.

Let's setup the hypotheses if the difference between our cancer types are statistically significant:

- **H₀ (Null):** There is no significant difference in mean survival time across the three cancer types (Lymphoma, Melanoma, and Retinoblastoma)
- **H₁ (Alternative):** At least one cancer type has a significantly different mean survival time compared to the others

```python
# Extract values from the summary table
lymphoma_mean = type_survival_mm.loc['mean', 'Lymphoma']
lymphoma_count = type_survival_mm.loc['count', 'Lymphoma']

melanoma_mean = type_survival_mm.loc['mean', 'Melanoma']
melanoma_count = type_survival_mm.loc['count', 'Melanoma']

retinoblastoma_mean = type_survival_mm.loc['mean', 'Retinoblastoma']
retinoblastoma_count = type_survival_mm.loc['count', 'Retinoblastoma']

# Calculate grand mean
grand_mean = ((lymphoma_mean * lymphoma_count) + (melanoma_mean * melanoma_count) + (retinoblastoma_mean * retinoblastoma_count)) / (lymphoma_count + melanoma_count + retinoblastoma_count)

print(f"Grand Mean (x̄): {grand_mean:.4f}")
print()

# Parameters for ANOVA
k = 3  # number of groups (cancer types)
n_total = lymphoma_count + melanoma_count + retinoblastoma_count  # total sample size
df1 = k - 1  # degrees of freedom between groups
df2 = n_total - k  # degrees of freedom within groups

print(f"Number of groups (k): {k}")
print(f"Total sample size (n): {n_total}")
print(f"Degrees of freedom between groups (df₁): {df1}")
print(f"Degrees of freedom within groups (df₂): {df2}")
print()

# Calculate SSG (Sum of Squares between Groups)
SSG = (lymphoma_count * (lymphoma_mean - grand_mean)**2 + 
       melanoma_count * (melanoma_mean - grand_mean)**2 + 
       retinoblastoma_count * (retinoblastoma_mean - grand_mean)**2)

print(f"Sum of Squares between Groups (SSG): {SSG:.4f}")

# Calculate SST (Total Sum of Squares)
# Using the orig survival data
all_survival_data = pd.concat([lymphoma_survival_df, melanoma_survival_df, retinoblastoma_survival_df])
SST = sum((x - grand_mean)**2 for x in all_survival_data)

print(f"Total Sum of Squares (SST): {SST:.4f}")

# Calculate SSE (Sum of Squares within groups/Error)
SSE = SST - SSG
print(f"Sum of Squares Error (SSE): {SSE:.4f}")

# Calculate Mean Squares
MSG = SSG / df1  # Mean Square between Groups
MSE = SSE / df2  # Mean Square Error

print(f"Mean Square between Groups (MSG): {MSG:.4f}")
print(f"Mean Square Error (MSE): {MSE:.4f}")

# Calculate F-statistic
F_statistic = MSG / MSE
print(f"F-statistic: {F_statistic:.4f}")

# Calculate p-value
from scipy.stats import f
p_value = 1 - f.cdf(F_statistic, df1, df2)
print(f"p-value: {p_value:.6f}")
print()

print("ANOVA RESULTS SUMMARY")
print(f"F-statistic: {F_statistic:.4f}")
print(f"p-value: {p_value:.6f}")
print(f"Alpha level: 0.05")

# Interpretation (formatted with claude sonnet 4.0)
print("\n" + "="*30)
print("STATISTICAL INTERPRETATION:")
print("="*30)
if p_value < 0.05:
    print("✅ RESULT: Reject null hypothesis (p < 0.05)")
    print("🔍 CONCLUSION: There IS a statistically significant difference")
    print("   in mean survival time across the three cancer types")
else:
    print("❌ RESULT: Fail to reject null hypothesis (p ≥ 0.05)")
    print("🔍 CONCLUSION: There is NO statistically significant difference")
    print("   in mean survival time across the three cancer types")
```

```python
# Create ANOVA table (formatted using Claude Sonnet 4.0)
anova_table = pd.DataFrame({
    'Source': ['Between Groups', 'Within Groups', 'Total'],
    'Sum of Squares': [SSG, SSE, SST],
    'df': [df1, df2, df1 + df2],
    'Mean Square': [MSG, MSE, ''],
    'F': [F_statistic, '', ''],
    'p-value': [p_value, '', '']
})

print("\n" + "="*20)
print("ANOVA TABLE:")
print("="*20)
print(anova_table.to_string(index=False))
```

#### **Statistical Inference 2:** *Testing if Stage at Diagnosis and Cancer Type have a significant effect on Survival Time*

To test if `Stage at Diagnosis` and `Cancer Type` have a significant difference on `Survival Time` we can utilize `Two-Way ANOVA`.<br>

Before we can proceed with this test, we have to check the following:
 - Our `Dependent Variable` (Survival Time) is `Normally-Distributed`
 - Our data exhibits `Homogeneity of Variance`

Let us first test if `Survival Time` is normally-distributed. For this, we can utilize the `Shapiro-Wilk Test`.<br>

This test will give us two important values:
 - `W value` : A value from 0 - 1. Value closer to 1 would  indicate that the data is normally distributed.
 - `p value` : Probability of obtaining the observed W value.

Hypotheses for Shapiro-Wilk Test:<br>

`Null Hypothesis`: 
    The dependent variable is normally-distributed.
    
`Alternative Hypothesis`:
    The dependent variable is not normally-distributed.

```python
from scipy.stats import shapiro

stat, p = shapiro(eye_cancer_df['Survival_Time_Months'])
print('Shapiro-Wilk test: Statistics=%.3f, p=%.3f' % (stat, p))
```

From the test above, we see a W value of 0.956 and a p value of 0.<br>

Even if the resulting W-value is close to 1, the p value is  significantly lower than 0.05, thus we `reject the null hypothesis`.<br>

Since our `dependent variable is not normally-distributed`, we cannot use Two-Way ANOVA. For this case, we would need to utilize a `non-parametric test`, this would not require our dependent variable to be normally-distributed.<br>

For our case, we shall use `Rank-Based ANOVA` as our non-parametric test. Which will rank our dependent variable and then do a `Two-Way ANOVA` on these said ranks.

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Copy and rank survival time
df = eye_cancer_df.copy()
df['Ranked_Survival'] = df['Survival_Time_Months'].rank()

# Model with interaction only
model = ols('Ranked_Survival ~ C(Stage_at_Diagnosis) * C(Cancer_Type)', data=df).fit()
anova_results = anova_lm(model)

# Display only the interaction row
interaction_result = anova_results.loc['C(Stage_at_Diagnosis):C(Cancer_Type)']
print(interaction_result)
```

From our result, we would only need the p-value to answer our hypothesis. From our results, we have a p-value of 0.102 which is greater than 0.05 thus we `fail to reject our null hypothesis`.<br>

> In conclusion, `Stage at Diagnosis` and `Cancer Type` does not significantly effect `Survival Time`.

```python
# Code for one-way ANOVA on Survival Time by Age Groups (Radiation Treatment Only)

# Separate survival times by age group for radiation treatment patients
age_groups_radiation = []
age_group_names = []
age_group_stats = []

for age_group in radiation_patients['Age_Group'].cat.categories:
    if age_group in radiation_patients['Age_Group'].values:  # Check if age group exists
        survival_data = radiation_patients[radiation_patients['Age_Group'] == age_group]['Survival_Time_Months']
        if len(survival_data) > 0:  # Only include non-empty groups
            age_groups_radiation.append(survival_data)
            age_group_names.append(age_group)
            
            # Calculate descriptive statistics
            mean_survival = survival_data.mean()
            median_survival = survival_data.median()
            std_survival = survival_data.std()
            age_group_stats.append({
                'Age_Group': age_group,
                'Count': len(survival_data),
                'Mean': mean_survival,
                'Median': median_survival,
                'Std': std_survival
            })
            
            print(f"{age_group}: n={len(survival_data)}, mean={mean_survival:.2f}, median={median_survival:.2f}, std={std_survival:.2f}")

print("\n" + "="*50)
print("ANOVA RESULTS")
print("="*50)

# Perform one-way ANOVA (only if we have at least 2 groups)
if len(age_groups_radiation) >= 2:
    f_statistic, p_value = stats.f_oneway(*age_groups_radiation)
    
    print(f"F-statistic: {f_statistic:.4f}")
    print(f"p-value: {p_value:.4f}")
    print(f"Alpha level: 0.05")
    print(f"Degrees of freedom between groups: {len(age_groups_radiation) - 1}")
    print(f"Degrees of freedom within groups: {sum(len(group) for group in age_groups_radiation) - len(age_groups_radiation)}")

    # Effect size (eta-squared)
    ss_between = f_statistic * (len(age_groups_radiation) - 1)
    ss_total = ss_between + (sum(len(group) for group in age_groups_radiation) - len(age_groups_radiation))
    eta_squared = ss_between / (ss_between + (sum(len(group) for group in age_groups_radiation) - len(age_groups_radiation)))
        
else:
    print("WARNING: Insufficient groups for ANOVA analysis")
    print(f" Found only {len(age_groups_radiation)} age groups with data")
    print(" ANOVA requires at least 2 groups")
```

> With a significance level of 0.05, One-Way ANOVA test for the Survival Time means of Age Groups in Radiation Treatment Patients yielded a p-value of 0.4623. <br>

```python
# Code for one-way ANOVA on Survival Time by Age Groups (Chemotherapy Treatment Only)

# Separate survival times by age group for chemotherapy treatment patients
age_groups_chemo = []
age_group_names_chemo = []
age_group_stats_chemo = []

for age_group in chemo_patients['Age_Group'].cat.categories:
    if age_group in chemo_patients['Age_Group'].values:  # Check if age group exists
        survival_data = chemo_patients[chemo_patients['Age_Group'] == age_group]['Survival_Time_Months']
        if len(survival_data) > 0:  # Only include non-empty groups
            age_groups_chemo.append(survival_data)
            age_group_names_chemo.append(age_group)
            
            # Calculate descriptive statistics
            mean_survival = survival_data.mean()
            median_survival = survival_data.median()
            std_survival = survival_data.std()
            age_group_stats_chemo.append({
                'Age_Group': age_group,
                'Count': len(survival_data),
                'Mean': mean_survival,
                'Median': median_survival,
                'Std': std_survival
            })
            
            print(f"{age_group}: n={len(survival_data)}, mean={mean_survival:.2f}, median={median_survival:.2f}, std={std_survival:.2f}")

print("\n" + "="*50)
print("ANOVA RESULTS")
print("="*50)

# Perform one-way ANOVA (only if we have at least 2 groups)
if len(age_groups_chemo) >= 2:
    f_statistic_chemo, p_value_chemo = stats.f_oneway(*age_groups_chemo)
    
    print(f"F-statistic: {f_statistic_chemo:.4f}")
    print(f"p-value: {p_value_chemo:.4f}")
    print(f"Alpha level: 0.05")
    print(f"Degrees of freedom between groups: {len(age_groups_chemo) - 1}")
    print(f"Degrees of freedom within groups: {sum(len(group) for group in age_groups_chemo) - len(age_groups_chemo)}")

else:
    print("WARNING: Insufficient groups for ANOVA analysis")
    print(f" Found only {len(age_groups_chemo)} age groups with data")
    print(" ANOVA requires at least 2 groups")
```

> With a significance level of 0.05, One-Way ANOVA test for the Survival Time means of Age Groups in Chemotherapy Treatment Patients yielded a p-value of 0.1926. <br>

## **Insights and Conclusion**

This project aimed to find out **what specific factors significantly correlate with the survival time of eye cancer patients**, through a combination of exploratory data analysis (EDA), clustering, association rule mining, and statistical inference. In the process, we were able to obtain critical insights and formulate a coherent data narrative, having been evident by empirical analysis.

The overall survival time for all patients demonstrated remarkable consistent, as indicated by the mean and median both being around **60 months**. This symmetry was also observed in the histogram and plots, indicating a roughly uniform distribution. This finding also held for the patients stratified by cancer types; their survival times were statistically similar. This was confirmed by ANOVA testing, which yielded a p-value of 0.8708, leading us to conclude that **cancer type does not substantially impact patient survival time**. We observed a repeating pattern when examining the stage at diagnosis. While Stage IV Retinoblastoma exhibited a median survival time that was unusually high, a non-parametric two-way ANOVA that included cancer type and stage as factors produced a p-value of 0.102, indicating no significant joint impact on survival.

Continuing to the observation to the **patients' age**, there were no notable correlation with survival time. The Pearson correlation coefficient was close to zero (0.0125), and distribution plots were not normally distributed, showing random uniform patterns instead. Moreover, one-way ANOVA analysis across treatment categories (Radiation, Chemotherapy, and Surgery) demonstrated that **age stratifications had no notable influence on survival time in any treatment subgroup**. All p-values derived from these analyses were above the 0.05 level of significance, indicating no statistical support for age being a determining factor in survival outcomes.

**Data mining techniques** such as **K-means clustering** revealed four main patient clusters (young/old paired with short/long survival), but the distribution of outcomes (active, deceased, in remission) remained consistent across clusters—hovering between **31% and 35%**—and offered little clinical insight. Trying to refine clustering by cancer types and stages, or by treatment type, still led to patterns that were statistically inconclusive. Similarly, **association rule mining** with the Apriori algorithm produced high confidence but low lift rules, suggesting that although some combinations of age and stage were frequently associated with longer survival, those associations were frail and most likely the result of the dataset's overall tilt toward favorable outcomes.

In conclusion, the study found that none of the variables—nor any group of cancer type, stage of diagnosis, type of treatment, or age—appeared to have a statistically significant effect on survival time. The extreme central tendency of a 60-month survival result in nearly all subgroups implies that the dataset might be **anomalously normalized** or synthetically balanced, a concern that aligns with the original disclaimer about its limitations. Therefore, while the statistical tests and machine learning methods were appropriately applied, the results point to the need for more detailed, clinically validated data to obtain genuinely actionable information for enhancing patient prognosis in clinical practice.
